```python
from bs4 import BeautifulSoup
import json
```

```python
data = ''
with open('./data/verblist.html', 'r') as infile:
    data = infile.read()

soup = BeautifulSoup(data, 'html.parser')
a_links = soup.find_all('a', class_='blue')
```

```python
main_table = soup.find_all('table', class_="vcr_table")[0]
```

```python
tester = main_table.find_all("table")[0].find_all("tr")[0]
def get_conj_page_and_translation(x):
    try:
        page = [_ for _ in x.find_all("a", class_="blue") if _['title'].find('conjugations') > -1][0]['href']
        contents = x.find_all("td", class_="i")[0].contents[0]    
        return contents, "https://www.laits.utexas.edu" + page
    except: 
        return None, None

```

```python
verbs = []
pages = []

for table in main_table.find_all("table"):
    for row in table.find_all("tr"):
        verb, page = get_conj_page_and_translation(row)
        if verb is not None or page is not None:
            verbs.append(verb)
            pages.append(page)

        
```
